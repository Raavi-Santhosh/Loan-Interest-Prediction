**Note :** ID, Monthly.Income and Inquiries in the last 6 months were only in Numerical and rest of the other are in categorical. So we need to convert them to numerical type 

Converting Amount requested and amount funded by investore numerical

```python
loan_data["Amount.Requested"] = pd.to_numeric(loan_data['Amount.Requested'],errors='coerce')
loan_data['Amount.Funded.By.Investors'] = pd.to_numeric(loan_data['Amount.Funded.By.Investors'],
                                                       errors='coerce')
```

Removing % sign in Interest rate and converting it to numeric

```python
loan_data['Interest.Rate']  = loan_data["Interest.Rate"].str.replace('%','')
loan_data['Interest.Rate' ] = pd.to_numeric(loan_data["Interest.Rate"],errors='coerce')
```

```python
loan_data.head()
```

Similarly removing % sign and converting that to numeric for Debt to income ratio variable

```python
loan_data['Debt.To.Income.Ratio'] = loan_data['Debt.To.Income.Ratio'].str.replace(
                                                "%",'')
loan_data['Debt.To.Income.Ratio'] = pd.to_numeric(loan_data["Debt.To.Income.Ratio"],errors='coerce')
```

```python
loan_data.info()
```

```python
loan_data.head()
```

Converting open credit lines and revolving credit balance 

```python
loan_data['Open.CREDIT.Lines'] = pd.to_numeric(loan_data['Open.CREDIT.Lines'],errors='coerce')
loan_data['Revolving.CREDIT.Balance'] = pd.to_numeric(loan_data['Revolving.CREDIT.Balance'],errors='coerce')
```

```python
loan_data.info()
```

Lets now see what can we do with loan.length variable

```python
loan_data["Loan.Length"].value_counts()
```

There are only 3 unique values for loan length out of with 36 and 60 months we are considering. Lets not create dummies for it. '.' value can be merged with 60 as only 1 value its holding on.

```python
loan_data['Loan.Length'] = np.where(loan_data['Loan.Length'] == "36 months",1,0)
loan_data['Loan.Length'].head()
```

Now make look of converting Loan purpose

```python
loan_data['Loan.Purpose'].value_counts()
```

There are 14 categories we can convert this type of feature into numeric in tow ways
    1. Groupping them by interest rate
    2. Creating dummies

***We group them by mean interest rate, if the performance is low we create dummies***

```python
round(loan_data.groupby("Loan.Purpose")["Interest.Rate"].mean())
```

car,educational,major_purchase has mean interest rate of 11

home_improvement,medical,vacation,wedding has mean interest rate of 12

creditcard, house, other, small business has mean interest rate of 13

debt consolidation, moving has mean interest rate of 14

renewable_enery has 10

```python
for i in range(loan_data.shape[0]):
    if loan_data["Loan.Purpose"][i] in ["car","educational","major_purchase"]:
        loan_data.loc[i,"Loan.Purpose"]="cem"
    if loan_data["Loan.Purpose"][i] in ["home_improvement","medical","vacation","wedding"]:
        loan_data.loc[i,"Loan.Purpose"]="hmvw"
    if loan_data["Loan.Purpose"][i] in ["credit_card","house","other","small_business"]:
        loan_data.loc[i,"Loan.Purpose"]="chos"
    if loan_data["Loan.Purpose"][i] in ["debt_consolidation","moving"]:
        loan_data.loc[i,"Loan.Purpose"]="dm"
```

Now create dummies for Loan purpose. Concatinate to loan_data and drop loan purpose variable

```python
lp_dummies = pd.get_dummies(loan_data["Loan.Purpose"],prefix='LP',drop_first=True)
```

```python
loan_data = pd.concat((loan_data,lp_dummies),axis='columns')
loan_data
```

```python
loan_data.drop("Loan.Purpose",axis="columns",inplace=True)
```

```python
loan_data.head()
```

Now just convert the home ownership column to numeric.

```python
loan_data["Home.Ownership"].value_counts()
```

As datapoints for Mortgage and Rent are equally distributed so adding OTHER and NONE to them randomly and create dummies

```python
#Adding Other to Mortgage and None to RENT
loan_data['Home.Ownership'].loc[loan_data['Home.Ownership'] == 'OTHER'] = "MORTGAGE"
```

```python
loan_data['Home.Ownership'].loc[loan_data['Home.Ownership'] == 'NONE'] = 'RENT'
```

```python
loan_data["Home.Ownership"].value_counts()
```

```python
lp_dummies = pd.get_dummies(loan_data['Home.Ownership'],prefix="HO",drop_first=True)
```

```python
loan_data = pd.concat((loan_data,lp_dummies),axis="columns")
```

```python
loan_data.drop('Home.Ownership',axis="columns",inplace=True)
```

Now we convert EMployement length to numeric by removing years,<,+

```python
loan_data['Employment.Length'] = loan_data['Employment.Length'].str.replace(" years","")
loan_data['Employment.Length'] = loan_data['Employment.Length'].str.replace("< ","")
loan_data['Employment.Length'] = loan_data["Employment.Length"].str.replace("+","")
loan_data['Employment.Length'] = loan_data['Employment.Length'].str.replace(" year","")
```

```python
loan_data['Employment.Length'] = pd.to_numeric(loan_data["Employment.Length"],errors='coerce')
```
